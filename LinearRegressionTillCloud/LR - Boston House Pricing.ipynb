```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import sklearn
from sklearn.datasets import load_boston

boston = load_boston()
boston
```

```python
df = pd.DataFrame(boston.data,columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD','TAX', 'PTRATIO', 'B', 'LSTAT'])
df.head()
```

```python
df["price"] = boston["target"]
df.head()
```

```python
df.isna().sum() ## No empty values
```

```python
## seeing how the feature data is distubuted

plt.figure(figsize=(25,15))
plotnumber = 1

for column in df:
    if plotnumber<=12:
        ax = plt.subplot(4,3,plotnumber) ## subplot(row,column, index of graph)
        sns.distplot(df[column],kde=True)
    plotnumber+=1
```

```python
## seeing how the label data is distubuted

plt.figure(figsize=(25,15))
plotnumber = 1

for column in df:
    if plotnumber<=12:
        ax = plt.subplot(4,3,plotnumber) ## subplot(row,column, index of graph)
        plt.scatter(df[column],df["price"])
    plotnumber+=1
```

```python
X = df.drop("price",axis=1)
y = df["price"]
X.head()
sam = (((X.iloc[0,1]) - (X["ZN"].mean()))/X["ZN"].std())
sam
```

```python
X.head()
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
X = X[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS','PTRATIO', 'B', 'LSTAT']]
X_scaled = scaler.fit_transform(X)
X_scaled
X.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = X_scaled ## vif can be calculated only on transformed parameters

vif = pd.DataFrame()

vif["VIF"] = [variance_inflation_factor(variables,i) for i in range((variables.shape[1]))]
    
vif["Features"] = X.columns
vif

## As we can see multi collinearity VIF more for TAX, preferring to delete the column and R2 is also not effected that much
```

```python
## Statistical analysis

import statsmodels.formula.api as smf
x = smf.ols(formula="price ~ CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+DIS+PTRATIO+B+LSTAT",data=df).fit()
print(x.rsquared)
print(x.rsquared_adj)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.25,random_state=355)

lm = LinearRegression()

lm.fit(x_train,y_train)
```

```python
def adjr2(x,y):
    r2 = lm.score(x,y)
    n =  x.shape[0]
    p = x.shape[1]
    adjr2 = 1 - (1-r2)*(n-1)/(n-p-1)
    return adjr2
```

```python
adjr2(x_train,y_train)
```

```python
adjr2(x_test,y_test)
```

```python
import pickle
filename = "finalmodel.pickle"
pickle.dump(lm,open(filename,"wb"))
```
